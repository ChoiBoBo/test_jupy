https://en.wikipedia.org/wiki/Natural_language_processing

위 링크를 클릭해서 들어갔다면, 위키백과 서버에 "위 주소에 해당하는 페이지를 보내달라"라고 요청을 보낸 셈이 됩니다. 같은 일은 브라우저 주소창에 주소를 입력하는 방식으로 할 수도 있습니다.

## 태그 이름을 통한 검색

`<h1>` 태그를 BeautifulSoup을 이용해 찾아보겠습니다.

```python
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/Natural_language_processing"
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser") # "lxml"
print(soup.h1)
```

마찬가지로, 다른 태그들도 검색할 수 있습니다.

```python
print(soup.title)
print(soup.b)

# 동일한 방법
# print(soup.find("title"))
# print(soup.find("b"))
```

`<h1>` 태그는 하나의 페이지에 한 개만 존재하기 때문에 위와 같은 방식으로 찾아도 상관 없습니다.

하지만 `<b>` 태그는 하나의 페이지 안에도 여러 개가 있을 수 있습니다. 위키백과 본문 중 굵은 표시가 된 텍스트에는 모두 `<b>` 태그가 달려 있습니다. 모든 `<b>` 태그를 가져오는 방법은 아래와 같습니다.

## 태그 속성을 이용한 검색

`<h1>` 태그 안에 보이는 class나 id 등을 태그의 속성이라고 합니다. 속성으로 태그를 검색하는 방법은 아래와 같습니다.

```python
print(soup.find(attrs={"class":"firstHeading"}))
```

```python
content_texts = soup.find(attrs={"class":"mw-parser-output"})

content_ls = []
for content in content_texts.find_all("p")[:2]:
    content_ls.append(content.text)
    
print("\n".join(content_ls))
```

## 링크 추출하기

한편 태그의 속성 자체가 중요한 정보인 경우도 있습니다. 태그 속성을 추출하는 방법은 아래와 같습니다.

```python
for tag in soup.find_all('a'):
    if 'href' in tag.attrs:
        if tag['href'].startswith('/wiki/'):
            print(tag['href'])
```

아래는 위키피디아의 여러 페이지를 방문하면서 데이터를 수집하는 코드입니다.

- parse_title: 제목을 수집합니다.
- parse_content: 본문을 수집합니다.
- parse_url: 링크를 수집합니다. 

`SEARCHING_WORDS`에 검색하고자 하는 단어를 넣은 뒤 실행시키면, 해당 페이지를 검색하고, 해당 페이지에 걸려있는 링크를 타고 넘어가 다시 데이터를 수집하는 것을 반복합니다.

```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd

def parse_title(soup):
    return soup.find(attrs={"id":"firstHeading"}).text

def parse_content(soup):
    tags = soup.find(attrs={"class":"mw-parser-output"}).find_all("p")
    contents = [tag.text for tag in tags]
    return ' '.join(contents).strip()
    
def parse_url(soup):
    urls = []
    for tag in soup.find_all('a'):
        if 'href' in tag.attrs:
            href = tag['href']
            if href.startswith('/wiki/') and ':' not in href:
                if "https://en.wikipedia.org"+href not in urls:
                    urls.append("https://en.wikipedia.org"+href)
    return urls

# 초기 URL requests
SEARCHING_WORDS = "Natural_language_processing"

queue = [f"https://en.wikipedia.org/wiki/{SEARCHING_WORDS}"]
data = []
i = 0
while queue:
    # 요청 보내기
    url = queue.pop(0)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    # 정보 추출하기
    title = parse_title(soup)
    content = parse_content(soup)
    data.append((title, content))
    queue.extend(parse_url(soup))
    
    i += 1
    if i > 5:
        break
```

```python
wiki_df = pd.DataFrame(data, columns=["title", "content"])
```
